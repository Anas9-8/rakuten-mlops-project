```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 50)
```

```python
# Load data
X_train = pd.read_csv('data/X_train.csv')
Y_train = pd.read_csv('data/Y_train.csv')

print(f"Training data shape: {X_train.shape}")
print(f"Training labels shape: {Y_train.shape}")
```

```python
# Display first few rows
print("First 5 products:\n")
X_train.head()
```

```python
# Check column info
print("Column information:")
print(X_train.info())
```

```python
# Check for missing values
print("Missing values per column:")
print(X_train.isnull().sum())
print(f"\nPercentage of missing descriptions: {(X_train['description'].isnull().sum() / len(X_train)) * 100:.2f}%")
```

```python
# Category distribution
category_counts = Y_train['prdtypecode'].value_counts()
print(f"Number of categories: {len(category_counts)}")
print(f"\nTop 10 categories:")
print(category_counts.head(10))
print(f"\nBottom 5 categories:")
print(category_counts.tail(5))
```

```python
# Plot category distribution
plt.figure(figsize=(14, 6))
category_counts.plot(kind='bar')
plt.title('Product Category Distribution', fontsize=16)
plt.xlabel('Product Type Code', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f"\n📊 Most common category: {category_counts.index[0]} with {category_counts.values[0]} products")
print(f"📊 Least common category: {category_counts.index[-1]} with {category_counts.values[-1]} products")
```
